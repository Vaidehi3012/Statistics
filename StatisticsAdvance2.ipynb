1. What is Hypothesis Testing?
  - Hypothesis testing is a formal procedure to decide whether data provide enough evidence to reject a stated assumption (the null hypothesis) about a population parameter.
  - You compute a test statistic from the sample, compare it to a sampling distribution under the null, and use a p-value or critical-value rule to accept or reject the null.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
  - The null hypothesis (H₀) is the default claim about a population parameter (e.g., μ = 50).
  - The alternative hypothesis (H₁ or Ha) is the competing claim (e.g., μ ≠ 50, μ > 50 or μ < 50).
  - Hypothesis tests evaluate evidence against H₀ in favor of H₁.

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
  - The significance level (α) is a threshold we choose before performing a hypothesis test — usually 0.05 (5%).
  - It represents the maximum probability of making a Type I error — rejecting the null hypothesis (H₀) when it is actually true.
  - Role in deciding the outcome:
  - Compute p-value: After running the test, we calculate a p-value (probability of observing data as extreme as ours under H₀).
  - Compare p-value with α:
  - If p ≤ α → Reject H₀ (evidence is strong enough against H₀).
  - If p > α → Fail to reject H₀ (not enough evidence to say H₀ is false).

4.  What are Type I and Type II errors? Give examples of each.
  - Type I Error (False Positive)
  - Definition: Rejecting the null hypothesis (H₀) when it is actually true.
  - Probability: Equal to the significance level (α).
  - Example:
  - Suppose a medical test is used to check if a patient has a disease.
  - H₀: The patient is healthy.
  - H₁: The patient has the disease.
  - Type I Error: The test says the patient is sick when they are actually healthy (false alarm).
  - Type II Error (False Negative)
  - Definition: Failing to reject the null hypothesis when it is actually false.
  - Probability: Denoted by β (related to the power of the test = 1 − β).
  - Example:
  - Using the same medical test:
  - Type II Error: The test says the patient is healthy when they actually have the disease (missed detection).

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
  - Z-Test
 - When to use:
  - Population standard deviation (σ) is known
  - Or sample size is large (n > 30) — by Central Limit Theorem, sample mean follows normal distribution
  - Distribution used: Standard Normal (Z) distribution
  - Example:
  - Testing if the average weight of packaged rice is 1 kg, when the population σ is known to be 0.05 kg.
  - T-Test
  - When to use:
  - Population standard deviation is unknown
  - Usually with small samples (n ≤ 30)
  - Uses sample standard deviation (s) to estimate σ
  - Distribution used: Student’s t-distribution (wider tails than normal, accounts for extra uncertainty)
  - Example:
  - Testing if the average marks of 15 students differ from 50, when population σ is unknown.

```python
#6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success
size = 10000  # number of random samples

# Generate binomial distribution
binomial_data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(binomial_data, bins=range(n+2), edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
'''7. Implement hypothesis testing using Z-statistics for a sample dataset inPython. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
'''
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

data = np.array(sample_data)
n = len(data)
sample_mean = np.mean(data)

# Hypotheses: H0: mu = 50 (null hypothesis), Ha: mu ≠ 50 (two-sided)
mu_0 = 50
sigma_known = 0.8  # Assume population standard deviation is known

# Z-statistic
z_stat = (sample_mean - mu_0) / (sigma_known / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
#8 Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate normal data
np.random.seed(42)  # for reproducibility
mu_true = 5.0       # true mean
sigma_true = 2.0    # true standard deviation
n = 120             # sample size

data = np.random.normal(loc=mu_true, scale=sigma_true, size=n)

# Calculate sample mean and standard error
sample_mean = np.mean(data)
sem = stats.sem(data)  # standard error of mean

# 95% confidence interval (t-based)
confidence = 0.95
df = n - 1
t_crit = stats.t.ppf((1 + confidence) / 2, df)
ci_lower = sample_mean - t_crit * sem
ci_upper = sample_mean + t_crit * sem

print(f"Sample Mean: {sample_mean:.4f}")
print(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")

# Plot histogram
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Simulation (n=120)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#9 Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def compute_and_plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot histogram.
    Z-score = (x - mean) / std
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=0)  # population standard deviation
    z_scores = (data - mean) / std

    # Plot histogram
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score (Standard Deviations from Mean)")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.legend()
    plt.show()

    return z_scores

# Example dataset
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=100)
z_scores = compute_and_plot_z_scores(sample_data)

print("First 10 Z-scores:", z_scores[:10])

'''Z-score Formula:

𝑍
=
𝑥
−
𝜇
𝜎
Z=
σ
x−μ
	​


where
𝑥
x = data point,
𝜇
μ = mean,
𝜎
σ = standard deviation.

Meaning:
Z-scores tell you how many standard deviations away a value is from the mean:

Z = 0 → exactly at the mean

Z = +1 → one standard deviation above mean

Z = -1 → one standard deviation below mean

Interpretation of Histogram:
The histogram shows most Z-scores clustering around 0. In normally distributed data:

~68% of Z-scores lie between -1 and +1

~95% lie between -2 and +2

This is useful for detecting outliers (Z-scores outside ±3 are usually considered extreme).'''
```
